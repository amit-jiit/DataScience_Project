#### Token-based matching
https://spacy.io/usage/rule-based-matching

```python
"""A token whose lowercase form matches “hello”, e.g. “Hello” or “HELLO”.
A token whose is_punct flag is set to True, i.e. any punctuation."""
"""each dictionary represents one token. 
If spaCy’s tokenization doesn’t match the tokens defined in a pattern,
the pattern is not going to produce any results"""

pattern = [{"LOWER": "building"},{"LOWER": "systems"},{"LOWER": "and"},{"LOWER": "algorithms"}]
                                                                                                                        
                                                                                                

"op - doesnot matter whether punctuation is there or not"
" ? -makes preceding symbol optional"
```

```python
"First, we initialize the Matcher with a vocab."
"The matcher must always share the same vocab with the documents it will operate on."
matcher = Matcher(nlp.vocab,)
"""We can now call matcher.add() with an ID and our custom pattern. 
The second argument lets you pass in an optional callback function to invoke on a successful match.
For now, we set it to None"""

matcher.add("matching process",None, pattern) # callback -none
```

```python
doc = nlp("Building Systems AND Algorithms")
```

```python
matches = matcher(doc)
matches
```

```python
for token in doc:
    print(token.text)
```

```python
"get the string value, from a list of (match_id, start, end) tuples"

for match_id,start,end in matches:
    string_id = nlp.vocab.strings[match_id]
    span = doc[start:end]
    print("hash value {} of string id {} ,".format(match_id,string_id),"matches from {} to {}".format(start,end),span.text)
```

### Combining NLP Model and Custom Rule

```python
import spacy
from spacy.matcher import Matcher
from spacy.tokens import Span
from spacy import displacy
```

```python
nlp = spacy.load('en_core_web_sm')

doc = nlp(' Dr. Amit chaired first board meeting at Google')

doc
```

```python
print([(ent.text, ent.label_) for ent in doc.ents])
```

```python
def add_title(doc):
    new_entities = []
    
    for ent in doc.ents:
        if ent.label_ == 'PERSON' and ent.start!= 0: # checking label(person) not org
            prev_token = doc[ent.start-1] # previous token of person
            
             # previous token could be any thing so we are matching token to our list
            if prev_token.text in ['Dr.','Dr','Mr','Mr.']:
                "modifing ent in doc that means token will now start from (start-1) position to include Dr,and updating label"
                
                new_ent = Span(doc,ent.start-1,ent.end,label=ent.label)
                new_entities.append(new_ent)
            # if token does not match then return prev or unmodified ent
            else:
                new_entities.append(ent)
    doc.ents=new_entities # to get out of loop
    return doc
            
```

```python
nlp =spacy.load('en_core_web_sm')
"add it after ner"
nlp.add_pipe(add_title, after = 'ner')
```

```python
"pass doc in added pipeline"
doc = nlp('Dr. Amit chaired first board meeting at Google')
```

```python
print([(ent.text, ent.label_) for ent in doc.ents])
```
