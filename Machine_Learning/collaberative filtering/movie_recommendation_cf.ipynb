```python
import pandas as pd
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
```

```python
# matching two string together
"FuzzyWuzzy is a library of Python which is used for string matching."
"Fuzzy string matching is the process of finding strings that match a given pattern."
from fuzzywuzzy import process
```

```python
movies = './dataset/movies.csv'
ratings = './dataset/ratings.csv'

df_movies = pd.read_csv(movies,usecols=['movieId','title'], dtype = {'movieID':'int32','title':'str'})
```

```python
df_movies.head()
```

```python
df_ratings = pd.read_csv(ratings,usecols=['userId','movieId','rating'],dtype = {'userId':'int32','movieID':'int32','rating':'float32'})
```

```python
df_ratings.head()
```

```python
## sparse matrix

movies_user = df_ratings.pivot(index='movieId',columns='userId',values='rating')
movies_user.head()
```

```python
movies_user=movies_user.fillna(0)
movies_user
```

```python
movies_user.isna().sum()
```

```python
# converting dataframe to matrix
movies_user_matrix = csr_matrix(movies_user.values)
movies_user_matrix
```

#### Buid Model

```python
# finding between distace between vectors
# euclidian distance
# manhattan distance
# cosine similiarity
# minkowski distance

model_knn = NearestNeighbors(metric = 'cosine', algorithm='brute',n_neighbors= 20)
```

```python
# train model
model_knn.fit(movies_user_matrix)
```

```python
# match movie name using fuzzywuzzy

process.extractOne('toy story',df_movies['title'])

# 0 index of movie in df_movies
```

```python
# recommender(movie_name) => list of movies

def recommender(movie_name,data,model,n_recommedation):
        model.fit(data)
        idx = process.extractOne(movie_name,df_movies['title'])[2] # 2 only for getting index of movie in df_movies
        print('movie_selected: ',df_movies['title'][idx],'index: ',idx)
        print('seraching for recommedation...')
        
        distances, indices = model.kneighbors(data[idx],n_neighbors= n_recommedation)
        
        #print(distances, indices)
        for i in indices:
            print(df_movies['title'][i].where(i!=idx)) # compare with other movie not itself
            
        
        



```

```python
recommender('Toy Story',movies_user_matrix,model_knn,20)
```
